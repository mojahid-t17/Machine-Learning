#Linear Regression
Linear regression is a statistical technique used to find the relationship between variables. In an ML context, linear regression finds the relationship between features and a label.

In Supervised Learning, we are trying to find a relationship between a dependent variable (y) and one or more independent variables (x). If the fitted model is cool, it can allow us to predict the value of y, given any values for x.


The Simple Linear Regression Setup exists when we are trying to relate 2 (or more) variables together. Linear Regression however only allows us to search for linear patterns in data.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/studentscores.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
x=df[['Hours']]  #double[] is used to Convert Series column  to DataFrame
y=df[['Scores']]
x.head()
y.head()
```

```python
#import linear regression model from sklearn
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
#train the ml model
reg.fit(x,y)
```

```python
# y=wx+b
w=reg.coef_
b=reg.intercept_
print(w,b)
# We can use this information to predict values of y!
```

```python
# Question 1: Predict the values for score, y when x is 4.5, 5.1 and 0
def predict_Y(x):
  return w*x +b
print(f"if the hour is 4.5 then y is: {predict_Y(4.5)}")
print(f"if the hour is 5.1 then y is: {predict_Y(5.1)}")
print(f"if the hour is 0 then y is: {predict_Y(0)}")
```

```python
reg.predict([[4.5],[5.1],[0]])
```

```python
df['predict_y']=reg.predict(x)
df
```

```python
plt.scatter(x.mean(),y.mean(),color='red')
plt.plot(x, reg.predict(x), color='green',label='y=wx+b')
plt.scatter(df['Hours'],df['Scores'])
plt.legend()
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python

print('The R2 score of the model on the test set is:', reg.score(x, y))
```
