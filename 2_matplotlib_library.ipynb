#Matplotlib is a low level graph plotting library in python that serves as a visualization utility.


```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting a simple function.
#create a basic line plot using Matplotlib.
price=[4500,5000,4000,3800,4800,5200,5500]
year=[2012,2013,2014,2015,2016,2017,2018]
plt.plot(year,price)
```

```python
# ploating from dataframe
df=pd.read_csv('/content/sharma-kohli.csv')
df
```

```python
plt.plot(df['index'],df['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
# Add a title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
# Add X-axis label
plt.xlabel('season')
# Add y-axis label
plt.ylabel('Scored Runs')
```

```python
#style level color ,
#add line style(supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted')
#add line width
plt.plot(df['index'],df['V Kohli'], color='red', linestyle='--',linewidth=4)
plt.plot(df['index'],df['RG Sharma'], color='green' ,linestyle='dashdot',linewidth=2)
```

```python
#marker : .: Point marker
# ,: Pixel marker
# o: Circle marker
# v: Triangle down marker
# ^: Triangle up marker
# <: Triangle left marker
# >: Triangle right marker
# s: Square marker
# *: Star marker
# +: Plus marker
# x: X marker
# D: Diamond marker
# d: Thin diamond marker
# p: Pentagon marker
# h: Hexagon marker
plt.plot(df['index'],df['V Kohli'], color='red', linestyle='--',linewidth=4, marker='d',markersize=10)
plt.plot(df['index'],df['RG Sharma'], color='green' ,linestyle='dashdot',linewidth=2, marker='o', markerfacecolor='blue')
```

#legend -> location
In Matplotlib, the location of the legend can be controlled using the loc parameter in the plt.legend() function.
 The loc parameter allows you to specify predefined positions or fine-tune the placement manually.
'best'	Automatically chooses the best location (default).
* 'upper right'-	Top-right corner.
* 'upper left'	-Top-left corner.
*'lower right'-	Bottom-right corner.
*'lower left'	-Bottom-left corner.
*'center'-	Center of the plot.
*'center right'- Center-right edge.
*'center left'	- Center-left edge.
*'upper center'- 	Centered at the top.
*'lower center'- 	Centered at the bottom.


```python
plt.plot(df['index'],df['V Kohli'], label='Virat ')
plt.plot(df['index'],df['RG Sharma'], label='Rohit')
# Add a title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
# Add X-axis label
plt.xlabel('season')
# Add y-axis label
plt.ylabel('Scored Runs')
plt.legend()
# plt.legend(loc='upper right')
```

```python
#grid
plt.plot(df['index'],df['V Kohli'], label='Virat ')
plt.plot(df['index'],df['RG Sharma'], label='Rohit')
# Add a title
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
# Add X-axis label
plt.xlabel('season')
# Add y-axis label
plt.ylabel('Scored Runs')
plt.legend()
# plt.legend(loc='upper right')
plt.grid()
```

# Scatter Plots
A scatter plot is a type of data visualization that represents data points on a two-dimensional plane using Cartesian coordinates. It is useful for identifying relationships, patterns, or trends between two variables.

```python

data=pd.read_csv('/content/batter.csv')
data
```

```python
#To create a scatter plot in Matplotlib, use the plt.scatter() function.
plt.scatter(data['avg'],data['strike_rate'])
# Add labels and title
plt.ylabel('Average run ')
plt.xlabel('Strike rate')
plt.title('avg and sR Scatter Plot')
```

```python
# color and marker
plt.scatter(data['avg'],data['strike_rate'], c='red', marker='+')
plt.ylabel('Average run ')
plt.xlabel('Strike rate')
plt.title('avg and sR Scatter Plot')
```

```python
#sine and cose wave subploting
x=np.arange(0,5*np.pi,0.1)
sin_x=np.sin(x)
cos_x=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,sin_x)
plt.subplot(2,1,2)
plt.plot(x,cos_x)
plt.show()
```
