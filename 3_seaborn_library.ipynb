#Seaborn is a Python visualization library based on Matplotlib, designed for statistical graphics and data visualization. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# 📚 Importing Required Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

🔹 1. Loading Example Datasets
Seaborn comes with built-in datasets.

```python
# Load dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
print(tips.info())
```

##🔹 2. Basic Plots in Seaborn
##📍 Scatter Plot
Used to show relationships between two numerical variables.

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex') #hue--> Colors the points based on the sex column (Male/Female).
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time') #style-->Uses different markers based on the time column (Lunch/Dinner).
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size') #size=> Adjusts the point size based on the size column (Number of people at the table).
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter') #The kind parameter selects the underlying axes-level function to use:
```

# **Line Plot**

```python
#import plotly for dataset
import plotly.express as px
```

```python
gap=px.data.gapminder()
temp_df=gap[gap['country']=='Bangladesh']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap[gap['country'].isin(['Bangladesh','China','Germany'])]
temp_df
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',hue='country',kind='line',style='continent',size='continent')
```

```python
#facetplot--> particular categorical column to multiple plot
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='day') #use colwrap=1,2,3.. for fixed column number
```

```python
#barplot
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.countplot(x="sex", data=tips)
plt.show()
```

```python
# plotting univariate histogram
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist',bins=2)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```
